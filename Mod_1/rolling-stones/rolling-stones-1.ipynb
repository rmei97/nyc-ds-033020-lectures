```python
def find_by_name(dict_data,album):
    # find someting in the reader
    for item in dict_data:
        x = collections.OrderedDict(item)
            # makes it all CAPS to compare
        if x['album'].upper() == album.upper():
            return x
        else:
            continue
    print("Not in the list")

    
find_by_name(dict_data,'London calling')
```

```python
def find_by_rank(dict_data,rank):
    for item in dict_data:
        x = collections.OrderedDict(item)
        if int(x['number']) == rank:
            return x
        else:
            continue
    print("This is not ranked")
```

```python
def find_by_year(dict_data,year):
    for item in dict_data:
        x = collections.OrderedDict(item)
        if int(x['year'])== year:
            return x
        else:
            continue
    print("None")
    
```

```python
def find_by_year_range(dict_data,year_1, year_2):
    list_of_albums=[]
    for item in dict_data:
        x = collections.OrderedDict(item)
        if int(x['year']) in range(year_1,year_2+1):
            list_of_albums.append(x['album'])
        else:
            continue
    return list_of_albums

            
```

```python
def find_by_rank_range(dict_data,rank_1, rank_2):
    list_of_albums=[]
    for item in dict_data:
        x = collections.OrderedDict(item)
        if int(x['number']) in range(rank_1,rank_2+1):
            list_of_albums.append(x['album'])
        else:
            continue
    return list_of_albums

            
```

```python
def all_titles(dict_data):
    list_of_albums=[]
    for item in dict_data:
        x = collections.OrderedDict(item)
        list_of_albums.append(x['album'])
    return list_of_albums
    
```

```python
all_titles(dict_data)
```

```python
def all_artists(dict_data):
    list_of_artists=[]
    for item in dict_data:
        x = collections.OrderedDict(item)
        list_of_artists.append(x['artist'])
    return list_of_artists
    
```

```python
def top_album(dict_data):
    list_of_artist= all_artists()
    set_artist= set(list_of_artist)
    current_top= list_of_artist[0]
    for artist in set_artist:
        if list_of_artist.count(current_top)< list_of_artist.count(artist):
            current_top= artist
        else:
            continue
    return current_top
        
```

```python
top_album(dict_data)
```

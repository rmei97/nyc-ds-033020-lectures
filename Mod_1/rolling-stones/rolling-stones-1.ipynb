```python
import csv
import collections 
import math
data = []
with open('data.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        data.append(row)
```

```python
def find_by_name(data,album):
    # find someting in the reader
    for item in data:
        x = collections.OrderedDict(item)
            # makes it all CAPS to compare
        if x['album'].upper() == album.upper():
            return x
    
find_by_name(data,'London calling')
```

```python
def find_by_rank(data,rank):
    for item in data:
        x = collections.OrderedDict(item)
        if int(x['number']) == rank:
            return x
```

```python
def find_by_year(data,year):
    for item in data:
        x = collections.OrderedDict(item)
        if int(x['year'])== year:
            return x
```

```python
def find_by_year_range(data,year_1, year_2):
    list_of_albums=[]
    for item in data:
        x = collections.OrderedDict(item)
        if int(x['year']) in range(year_1,year_2+1):
            list_of_albums.append(x['album'])
    return list_of_albums

            
```

```python
def find_by_rank_range(data,rank_1, rank_2):
    list_of_albums=[]
    for item in data:
        x = collections.OrderedDict(item)
        if int(x['number']) in range(rank_1,rank_2+1):
            list_of_albums.append(x['album'])
    return list_of_albums
```

```python
def all_titles(data):
    list_of_albums=[]
    for item in data:
        x = collections.OrderedDict(item)
        list_of_albums.append(x['album'])
    return list_of_albums
    
```

```python
def all_artists(data):
    list_of_artists=[]
    for item in data:
        x = collections.OrderedDict(item)
        list_of_artists.append(x['artist'])
    return list_of_artists
    
```

```python
def top_album(data):
    list_of_artist= all_artists(data)
    set_artist= set(list_of_artist)
    current_top = list_of_artist[0]
    for artist in set_artist:
        if list_of_artist.count(current_top)< list_of_artist.count(artist):
            current_top= artist
        else:
            continue
    return current_top
        
```
